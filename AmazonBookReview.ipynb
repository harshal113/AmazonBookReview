```python
import pandas as pd
```

```python
abr = pd.read_csv(r"C:\Users\admin\Downloads\AmazonBookReviews.csv",encoding='latin-1')
```

```python
abr.head()
```

```python
df = pd.DataFrame()
df['X']=abr.review
df['Y']=abr.label
```

```python
df.Y.unique()
```

```python
df=df.replace({'pos':1,'neg':0})
```

```python
df.head()
```

```python
df.X=df.X.str.lower()
```

```python
import nltk
from nltk.corpus import stopwords
```

```python
stopwords.words('english')
l1=list(stopwords.words('english'))
```

```python
import string
```

```python

def text_process(mess):
    """
    1.remove the punctuation 
    2.remove the stopwords
    3.return the list of clean textwords
    """
    nopunc=[char for char in mess if char not in string.punctuation]
    nopunc="".join(nopunc)
    return[word for word in nopunc.split() if word not in l1]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
bow_transfomer=CountVectorizer(analyzer=text_process).fit(df.X)
```

```python
tdm=bow_transfomer.transform(df["X"])
```

```python
tdm.shape
```

```python
from sklearn.model_selection import train_test_split

tdm_train,tdm_test,y_train,y_test=train_test_split(tdm,df.Y,test_size=.2)
```

```python
#naive bayes

from sklearn.naive_bayes import MultinomialNB
nb_abr = MultinomialNB()

nb_abr.fit(tdm_train, y_train)
```

```python
pred_nb_abr = nb_abr.predict(tdm_test)
pred_nb_abr
```

```python
from sklearn.metrics import confusion_matrix ,classification_report
```

```python
confusion_matrix(y_test,pred_nb_abr)
```

```python
print(classification_report(y_test,pred_nb_abr))
```

```python
#logistic

from sklearn.linear_model import LogisticRegression
logreg_abr = LogisticRegression()
```

```python
logreg_abr.fit(tdm_train,y_train)
```

```python
pred_log_abr = logreg_abr.predict(tdm_test)
pred_log_abr
```

```python
confusion_matrix(y_test,pred_log_abr)
```

```python
print(classification_report(y_test,pred_log_abr))
```

```python
#decision tree
from sklearn.tree import DecisionTreeClassifier
dt_abr = DecisionTreeClassifier()

dt_abr.fit(tdm_train, y_train)
```

```python
pred_dt_abr = dt_abr.predict(tdm_test)
pred_dt_abr
```

```python
confusion_matrix(y_test,pred_dt_abr)
```

```python
print(classification_report(y_test,pred_dt_abr))
```

```python
#random forrest

from sklearn.ensemble import RandomForestClassifier
rf_abr = RandomForestClassifier()
```

```python
rf_abr.fit(tdm_train, y_train)
```

```python
pred_rf_abr = rf_abr.predict(tdm_test)
pred_rf_abr
```

```python
confusion_matrix(y_test,pred_rf_abr)
```

```python
print(classification_report(y_test,pred_rf_abr))
```

```python
abr = pd.read_csv(r"C:\Users\admin\Downloads\AmazonBookReviews.csv",encoding='latin-1')
```

```python
abr.head()
```

```python
abr=abr.loc[:,['review']]
```

```python
abr.review=abr.review.str.lower()
```

```python
#textblob
from textblob import TextBlob
```

```python
score_list=[]
for i in range (0,abr.shape[0]):
    score=TextBlob(abr.review[i])
    score_list.append(score.sentiment[0])

score_list
```

```python
abr['score']=score_list
```

```python
abr[abr.score<=0].shape
```

```python
abr[abr.score>=0].shape
```

```python
# vader

from nltk.sentiment.vader import SentimentIntensityAnalyzer

sent=SentimentIntensityAnalyzer()
```

```python
score_list=[]
for i in range (0,abr.shape[0]):
    score=sent.polarity_scores(abr.review[i])
    score_list.append(score['compound'])

score_list
```

```python
abr['compound_score']=score_list
```

```python
abr[abr.compound_score<=0].shape
```

```python
abr[abr.compound_score>=0].shape
```
